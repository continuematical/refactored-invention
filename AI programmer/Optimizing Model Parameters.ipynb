```python
import torch
from torch import nn
from torch import optim
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
```

```python
training_data = datasets.FashionMNIST(
    root="../data",
    train=True,
    download=True,
    transform=ToTensor()
)

test_data = datasets.FashionMNIST(
    root="../data",
    train=False,
    download=True,
    transform=ToTensor()
)

train_dataloader=DataLoader(training_data, batch_size=64)
test_dataloader=DataLoader(test_data, batch_size=64)
```

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten=nn.Flatten()
        self.linear=nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512,512),
            nn.ReLU(),
            nn.Linear(512,10)
        )
        
    def forward(self, x):
        x=self.flatten(x)
        x=self.linear(x)
        return x
```

```python
model=NeuralNetwork()
```

```python
learning_rate=1e-3
batch_size=64
epochs=5

# 损失函数和优化器
loss_cn=nn.CrossEntropyLoss()
optimizer=optim.SGD(model.parameters(), lr=0.01)
```

```python
# 编写训练函数和测试函数
def train_loop(dataLoader, model, loss_cn, optimizer):
    size=len(dataLoader.dataset)
    # 开始训练模式
    model.train()
    for batch, (X, y) in enumerate(dataLoader):
        pred=model(X)
        loss=loss_cn(pred, y)
        # 反向传播
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
        if batch % 100 == 0:
            loss, current=loss.item(), batch*batch_size+len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def test_loop(dataLoader, model, loss_cn):
    # 将模型转换为评价模式
    model.eval()
    size=len(dataLoader.dataset)
    num_batches=len(dataLoader)
    test_loss, correct=0,0
    
    with torch.no_grad():
        for X, y in dataLoader:
            pred=model(X)
            test_loss+=loss_cn(pred, y).item()
            correct+=(pred.argmax(1)==y).type(torch.float).sum().item()
            
    test_loss/=num_batches
    correct/=size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

epochs = 10
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_dataloader, model, loss_cn, optimizer)
    test_loop(test_dataloader, model, loss_cn)
print("Done!")
```
